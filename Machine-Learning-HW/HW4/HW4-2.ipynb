```python
#讀入資料
import pandas as pd
Data = pd.read_csv('HW4_creditcard.csv')
Data
```

```python
from sklearn.model_selection import train_test_split
trainData, testData, a, b = train_test_split(Data, range(Data.shape[0]), test_size=0.3, random_state=None)
print("Train Data")
trainData
```

```python
print("Test Data")
testData
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
#直接訓練
def fromDataframeToPrediectResult(comment, trainData, testData):
    trainLabel, testLabel = list(trainData['Class']), list(testData['Class'])
    trainData = np.array(trainData.drop(['Class'], axis = 1))
    testData = np.array(testData.drop(['Class'], axis = 1))
    #訓練
    Model = DecisionTreeClassifier(random_state=0).fit(trainData, trainLabel) if comment == "Decision" else MLPClassifier(random_state=1).fit(trainData, trainLabel)
    #預測
    ModelResult = Model.predict(testData)
    #統計結果
    print(comment, "Data with")
    
    tn, fp, fn, tp = confusion_matrix(testLabel, ModelResult).ravel()
    accuracy = (tp+tn)/(tn+fp+fn+tp)
    recall = tp/(tp+fn)
    precision = tp/(tp+fp)
    F1 = 2/(1/recall + 1/precision)
    print("accuracy:",round(accuracy*100,2),"%", end = ' ')
    print("recall:",round(recall*100,2),"%", end = ' ')
    print("precision:",round(precision*100,2),"%", end = ' ')
    print("F1-Score:",round(F1*100,2),"%")

fromDataframeToPrediectResult("Neural Network", trainData, testData)

fromDataframeToPrediectResult("Decision", trainData, testData)
```

```python
#觀察class 1和class2的數量
print("Orignal Data Label count:")
label = list(int(i) for i in Data['Class'])
print("# Class 0:",label.count(0), "# Class 1:", label.count(1))
```

```python
#改變權重
def fromDataframeToPrediectResultWithChangeArgument(comment, trainData, testData):
    trainLabel, testLabel = list(trainData['Class']), list(testData['Class'])
    trainData = np.array(trainData.drop(['Class'], axis = 1))
    testData = np.array(testData.drop(['Class'], axis = 1))
    #訓練
    Model = DecisionTreeClassifier(random_state=0,class_weight="balanced").fit(trainData, trainLabel) if comment == "Decision" else MLPClassifier(random_state=1,max_iter = 300,hidden_layer_sizes=(50,)).fit(trainData, trainLabel)
    #預測
    ModelResult = Model.predict(testData)
    #統計結果
    print(comment, "Data with")
    
    tn, fp, fn, tp = confusion_matrix(testLabel, ModelResult).ravel()
    accuracy = (tp+tn)/(tn+fp+fn+tp)
    recall = tp/(tp+fn)
    precision = tp/(tp+fp)
    F1 = 2/(1/recall + 1/precision)
    print("accuracy:",round(accuracy*100,2),"%", end = ' ')
    print("recall:",round(recall*100,2),"%", end = ' ')
    print("precision:",round(precision*100,2),"%", end = ' ')
    print("F1-Score:",round(F1*100,2),"%")

fromDataframeToPrediectResultWithChangeArgument("Neural Network", trainData, testData)

fromDataframeToPrediectResultWithChangeArgument("Decision", trainData, testData)
```

```python
#使用SMOTE套件使資料平衡
from imblearn.over_sampling import SMOTE
resampledData, resampledLabel = SMOTE().fit_resample(Data, Data['Class'])
resampledData
```

```python
#平衡資料訓練結果
print("Resample Data Label count:")
print("# Class 0:",list(resampledLabel).count(0), "# Class 1:", list(resampledLabel).count(1), end = '\n\n')

resampledtrainData, resampledtestData, a, b = train_test_split(resampledData, range(resampledData.shape[0]), test_size=0.3, random_state=None)

fromDataframeToPrediectResult("Neural Network", resampledtrainData, resampledtestData)

fromDataframeToPrediectResult("Decision", resampledtrainData, resampledtestData)
```

# 比較訓練結果
一開始沒有經過處理的資料進行訓練可以發現accuracy 雖然很高，但是recall的表現非常的差，F1的分數只有0.159而已。

因此我決定先改變兩個模型的訓練參數:

Neural Network的class_weight改成"balanced"；Decision Tree的hide layer減少為(50,)，max_iter增加為300。

可以發現accuracy稍微升高，另外recall、precision和F1也提高不少。

最後則是透過SMOTE套件來使class 1和class 2的分布結果更加平衡，訓練的結果可以發現，accuracy、recall、precision、F1的分數都非常高。
