```python
#將訓練與測試資料轉為np.array 並resize至128*128
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from random import random as ErrorClassfication

trainSize = 420
AfricanImg,AsianImg,TestImg = list(), list(), list()
for i in range(1,trainSize+1):
    try:
        AfricanImg.append(np.array(Image.open('train/African/af_tr'+str(i)+'.jpg').resize((128,128))))
    except:
        AfricanImg.append(np.array(Image.open('train/African/af_tr'+str(i)+'.jpeg').resize((128,128))))
    try:
        AsianImg.append(np.array(Image.open('train/Asian/as_tr'+str(i)+'.jpg').resize((128,128))))
    except:
        AsianImg.append(np.array(Image.open('train/Asian/as_tr'+str(i)+'.jpeg').resize((128,128))))

TestSize = 188
for i in range(TestSize):
    TestImg.append(np.array(Image.open('test/e'+('00' if i < 10 else('0' if i < 100 else ''))+str(i)+'.jpg').resize((128,128))))

trainError, testError = list(),list()
trainData ,testData = np.empty(shape=(2*trainSize,128,128,3)), np.empty(shape=(TestSize,128,128,3))
for i in range(trainSize):
    try:
        trainData[2*i] = np.array(AfricanImg[i])
    except ValueError:
        trainError.append(2*i)
    try:
        trainData[2*i+1] = np.array(AsianImg[i])
    except ValueError:
        trainError.append(2*i+1)
for i in range(TestSize):
    try:
        testData[i] = np.array(TestImg[i])
    except:
        testError.append(i)
        
trainData = np.delete(trainData,trainError,axis = 0)
testData = np.delete(testData,testError, axis = 0)
print("Here")
```

```python
#建立訓練模型
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Conv2D, Flatten, MaxPooling2D
model = Sequential([
    Conv2D(
        batch_input_shape=(None,128,128,3),
        filters = 32,
        kernel_size=(5,5),
        strides=1,
        padding='same',
        name="Conv2D-1",
    ),
    Activation('relu'),
    MaxPooling2D(
        pool_size=(2,2),
        strides=2,
        padding='same',
        name="MaxPool-1",
    ),
    Conv2D(
        filters = 64,
        kernel_size=(5,5),
        strides=1,
        padding='same',
        name="Conv2D-2",
    ),
    Activation('relu'),
    MaxPooling2D(
        pool_size=(2,2),
        strides=2,
        padding='same',
        name="MaxPool-2",
    ),
    Conv2D(
        filters = 128,
        kernel_size=(5,5),
        strides=1,
        padding='same',
        name="Conv2D-3",
    ),
    Activation('relu'),
    MaxPooling2D(
        pool_size=(2,2),
        strides=2,
        padding='same',
        name="MaxPool-3",
    ),
    Conv2D(
        filters = 256,
        kernel_size=(5,5),
        strides=1,
        padding='same',
        name="Conv2D-4",
    ),
    Activation('relu'),
    MaxPooling2D(
        pool_size=(2,2),
        strides=2,
        padding='same',
        name="MaxPool-4",
    ),
    Flatten(),
    Dense(8192),
    Activation('relu'),
    Dense(1024),
    Activation('relu'),
    Dense(2),
    Activation('softmax'),
])
model.summary()
```

```python
#訓練
import tensorflow
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Conv2D, Flatten, MaxPooling2D
from tensorflow.keras.optimizers import SGD

trainLabel = [[0,1],[1,0]]*420
for i in trainError:
    trainLabel.pop(i)

if input("Train Model(y/n)?") == 'y':
    model.compile(loss='binary_crossentropy',optimizer = 'adam',metrics=['accuracy'])
    model.fit(trainData, np.array(trainLabel), epochs=20, batch_size=128)
    #儲存權重
    model.save_weights('elephant.h5')
```

```python
#測試資料
result = list(model.predict(testData))
for i in testError:
    result.insert(i,[ErrorClassfication(),0])
```
