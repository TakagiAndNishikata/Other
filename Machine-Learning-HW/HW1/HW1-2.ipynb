```python
import pandas as pd
import numpy as np
orignalData = pd.read_csv('HW1_data.csv').sort_values(by = ['0', '1'])
df = pd.DataFrame(dict())
df['Name'] = None
df['Sex'] = None
df['Age'] = None
df['HR'] = None
df['Height'] = None
df['Weight'] = None
df['BP'] = None
count = 0
temp = ''
for i in range(orignalData.shape[0]):
    if orignalData.iloc[i,0] != temp:
        df.loc[str(count),orignalData.iloc[i,1]] = float(orignalData.iloc[i,2]) if orignalData.iloc[i,1] != 'Sex' else orignalData.iloc[i,2]
        temp = df.loc[str(count),'Name'] = orignalData.iloc[i,0]
        count += 1
    else:
        df.loc[str(count-1),orignalData.iloc[i,1]] = float(orignalData.iloc[i,2]) if orignalData.iloc[i,1] != 'Sex' else orignalData.iloc[i,2]
df
```

```python
print("取得平均")
def average(x):
    try:
        x = x.astype('float')
        sum = num = 0
        for i in x:
            if not np.isnan(i):
                sum += i
                num += 1
        return round(sum/num,1)
    except ValueError:
        return ''
df.loc['Average'] = df.apply(average,raw = True,axis = 0)
print(df)
df.drop('Average',inplace = True)
```

```python
print("各項數值之最大值的姓名")
print("Age:", df.sort_values(by = ['Age'],na_position = 'last', ascending = False).iloc[0,0],int(df.sort_values(by = ['Age'],na_position = 'last', ascending = False).iloc[0,2]), "years old")
print("HR:", df.sort_values(by = ['HR'],na_position = 'last', ascending = False).iloc[0,0],int(df.sort_values(by = ['HR'],na_position = 'last', ascending = False).iloc[0,3]))
print("Height:", df.sort_values(by = ['Height'],na_position = 'last', ascending = False).iloc[0,0],int(df.sort_values(by = ['Height'],na_position = 'last', ascending = False).iloc[0,4]),"cm")
print("Weight:", df.sort_values(by = ['Weight'],na_position = 'last', ascending = False).iloc[0,0],df.sort_values(by = ['Weight'],na_position = 'last', ascending = False).iloc[1,0],df.sort_values(by = ['Weight'],na_position = 'last', ascending = False).iloc[2,0],int(df.sort_values(by = ['Weight'],na_position = 'last', ascending = False).iloc[0,5]),"Kg")
print("BP:", df.sort_values(by = ['BP'],na_position = 'last', ascending = False).iloc[0,0], int(df.sort_values(by = ['BP'],na_position = 'last', ascending = False).iloc[0,6]))
```
